```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/fut_prices_raw.csv")
```

```python
#amount of data with a price of 0
df[df["price"] == 0]
```

```python
#replacing prices with values of zero with NaN as they are missing data
df["price"] = df["price"].replace(0, np.nan)
#quick check
df[df["price"] == 0]
```

```python
#As we are dealing with time series data, we need to make sure it is sorted properly
df = df.sort_values(["player_id", "date"])
df.head()
```

```python
#this will fill the missing values with the data of the previous time
df["price"] = df.groupby("player_id")["price"].ffill()
```

```python
#this will fill the missing values with the data of the future time (used when there are days missing at the beginning)
df["price"] = df.groupby("player_id")["price"].bfill()
```

```python
df["pct_change"] = df.groupby("player_id")["price"].pct_change()
df["pct_change"] = df["pct_change"].clip(-1, 1)
```
